```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df1 = np.load('DGM_relative_error_4D_Dirichlet_nonlinear_preset_boundary.npy')
df2 = np.load('DRM_relative_error_4D_Dirichlet_nonlinear_preset_boundary.npy')
df3 = np.load('DGM_relative_error_4D_Dirichlet_nonlinear_swish.npy')
df4 = np.load('DRM_relative_error_4D_Dirichlet_nonlinear_swish.npy')
```

```python
epoch = [i for i in range(len(df3))]
```

```python
DGM_epoch1 = list(epoch[:])[:9999:250]
DGM_error1 = list(df1[:])[:9999:250]

DRM_epoch2 = list(epoch[:])[:9999:250]
DRM_error2 = list(df2[:])[:9999:250]

DGM_epoch3 = list(epoch[:])[::250]
DGM_error3 = list(df3[:])[::250]

DRM_epoch4 = list(epoch[:])[::250]
DRM_error4 = list(df4[:])[::250]
```

```python
plt.plot(DGM_epoch1, DGM_error1, color = "b", linestyle= "-", marker = "x", linewidth = 0.5, label = "DGM without penalty")
plt.plot(DRM_epoch2, DRM_error2, color = "r", linestyle= "-", marker = ".", linewidth = 1.5, label = "DRM without penalty")
plt.plot(DGM_epoch3, DGM_error3, color = "g", linestyle= "-.", marker = "+", linewidth = 0.5, label = "DGM + penalty")
plt.plot(DRM_epoch4, DRM_error4, color = "k", linestyle= "-.", marker = "^", linewidth = 1.5, label = "DRM + penalty")
plt.xlabel("epoch")
plt.ylabel("error")
plt.legend(loc = 'upper left', bbox_to_anchor = (0.5, 0.95))
plt.savefig("nonlinear_dirichlet_error_4d.eps", dpi = 120)
plt.show() 
```
