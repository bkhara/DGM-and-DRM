```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df1 = np.load('DGM_relative_error_4D_Dirichlet.npy')
df2 = np.load('DRM_relative_error_4D_Dirichlet.npy')
```

```python
epoch = [i for i in range(len(df1))]
```

```python
DGM_epoch1 = list(epoch[:])[::800]
DGM_error1 = list(df1[:])[::800]
DRM_epoch1 = list(epoch[:])[::800]
DRM_error1 = list(df2[:])[::800]
```

```python
plt.plot(DGM_epoch1, DGM_error1, color = "b", linestyle= "-.", marker = "x", linewidth = 0.5, label = "DGM")
plt.plot(DRM_epoch1, DRM_error1, color = "r", linestyle= "-", marker = ".", linewidth = 1.5, label = "DRM")
plt.xlabel("epoch")
plt.ylabel("error")
plt.legend(loc = 'upper left', bbox_to_anchor = (0.7, 0.95))
plt.savefig("dirichlet_error_4d.eps", dpi = 120)
plt.show() 
```
